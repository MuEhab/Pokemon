```python
legendary_counts = df["Legendary"].value_counts()
labels = legendary_counts.index
values = legendary_counts.values 
plt.figure(figsize=(6, 6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Legendary PokÃ©mon")
plt.axis('equal')
plt.show()
```

```python
condition = df['Name'] == "Mew"
df = df[~condition]
df = df.reset_index(drop=True)
df
```

```python
pokemon_less_than_500 = df[df["Total"] < 500]
pokemon_greater_than_500 = df[df["Total"] >= 500]
count_less_than_500 = len(pokemon_less_than_500)
count_greater_than_500 = len(pokemon_greater_than_500)
pokemon_counts = [count_less_than_500, count_greater_than_500]
labels = ["Total < 500", "Total >= 500"]
plt.figure(figsize=(6, 6))
plt.pie(pokemon_counts, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of PokÃ©mon by Total Score")
plt.axis('equal') 
plt.show()
```
